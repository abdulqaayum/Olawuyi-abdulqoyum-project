```python
def calculate_metrics(df):
    df['mortality_rate'] = (df['total_deaths'] / df['total_confirmed']) * 100
    df['recovery_rate'] = (df['total_recovered'] / df['total_confirmed']) * 100
    df['tests_per_case'] = df['total_tests'] / df['total_confirmed']
    return df
```

```python
def analyze_trends(df, date_col, metric_col, freq):
    df[date_col] = pd.to_datetime(df[date_col])
    trends = df.groupby(pd.Grouper(key=date_col, freq=freq))[metric_col].sum().reset_index()
    return trends
```

```python
def compare_regions(df, group_col, metric_col):
    comparison = (
        df.groupby(group_col)[metric_col]
        .sum()
        .reset_index()
        .sort_values(by=metric_col, ascending=False)
    )
    return comparison
```

```python
def visualize_trends(df, x_col, y_col, hue_col=None, title=None):
    plt.figure(figsize=(12, 6))
    sns.lineplot(data=df, x=x_col, y=y_col, hue=hue_col)
    plt.title(title or f'{y_col} Trends')
    plt.xlabel(x_col)
    plt.ylabel(y_col)
    plt.show()
```

```python
continent_comparison = compare_regions(df, group_col='continent', metric_col='total_confirmed')
print(continent_comparison)
```

```python
continent_comparison.plot(kind='bar', x='continent', y='total_confirmed', legend=False, figsize=(10, 6))
plt.title("Total Confirmed Cases by Continent")
plt.ylabel("Total Confirmed Cases")
plt.xlabel("Continent")
plt.xticks(rotation=45)
plt.show()
```

```python
df = calculate_metrics(df)
mortality_comparison = compare_regions(df, group_col='continent', metric_col='mortality_rate')
print(mortality_comparison)
```

```python
recovery_comparison = compare_regions(df, group_col='continent', metric_col='recovery_rate')
recovery_comparison.plot(kind='bar', x='continent', y='recovery_rate', legend=False, figsize=(10, 6))
plt.title("Recovery Rates by Continent")
plt.ylabel("Recovery Rate (%)")
plt.xlabel("Continent")
plt.xticks(rotation=45)
plt.show()
```
